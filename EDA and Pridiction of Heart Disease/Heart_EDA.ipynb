```python
from sklearn.model_selection import train_test_split
train=data.iloc[:,:-1]
test=data.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=0.2,random_state=1)

```

```python
print("x_train len",len(x_train))
print("x_test len",len(x_test))
print("y_train len",len(y_train))
print("x_test len",len(y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
accuracy_list_train=[]
accuracy_list_test=[]
dt=DecisionTreeClassifier()
splits=[2,10,30,50,100,200,300,700]

for i in splits:
  dt=DecisionTreeClassifier(min_samples_split=i,random_state=1).fit(x_train,y_train)
  prediction_x_train=dt.predict(x_train)
  prediction_x_test=dt.predict(x_test)

  training_accuracy=accuracy_score(prediction_x_train,y_train)
  testing_accuracy=accuracy_score(prediction_x_test,y_test)
  accuracy_list_test.append(testing_accuracy)
  accuracy_list_train.append(training_accuracy)


```

```python
training_accuracy=accuracy_score(prediction_x_train,y_train)
testing_accuracy=accuracy_score(prediction_x_test,y_test)
print(testing_accuracy)
print(training_accuracy)
```

```python
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_test)
plt.xticks(ticks = range(len(splits)),labels=splits)
plt.title("Train x validation metrics")
plt.xlabel('samples split')
plt.ylabel('accuracy')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

depths = [2, 4, 8, 16, 32, 64]
accuracy_list_train=[]
accuracy_list_test=[]

for i in depths:
  rf=RandomForestClassifier(min_samples_split=i,random_state=1).fit(x_train,y_train)
  prediction_x_train=rf.predict(x_train)
  prediction_x_test=rf.predict(x_test)

  training_accuracy=accuracy_score(prediction_x_train,y_train)
  testing_accuracy=accuracy_score(prediction_x_test,y_test)
  accuracy_list_test.append(testing_accuracy)
  accuracy_list_train.append(training_accuracy)
```

```python
plt.plot(accuracy_list_test)
plt.plot(accuracy_list_train)
plt.title("Train x validation metrics")
plt.xlabel('max depth')
plt.ylabel('accuracy')
plt.legend(['Train', 'Validation'])
plt.xticks(ticks = range(len(depths)),labels=depths)
plt.show()
```

```python
training_accuracy=accuracy_score(prediction_x_train,y_train)
testing_accuracy=accuracy_score(prediction_x_test,y_test)
print(testing_accuracy)
print(training_accuracy)
```

```python
random_forest_model = RandomForestClassifier(n_estimators = 500,max_depth=16, min_samples_split=10).fit(x_train,y_train)
print(f"Matrices train: {accuracy_score(random_forest_model.predict(x_train),y_train):.4f}")
print(f"Metrices test: {accuracy_score(random_forest_model.predict(x_test),y_test):.4f}")
```
